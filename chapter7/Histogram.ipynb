```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
cat = cv2.imread('cat.jpg',0)
plt.imshow(cat,cmap='gray')
plt.show()
```

```python
plt.hist(cat.ravel(),256,[0,256]);
plt.show()
```

```python
cat = cv2.imread('cat.jpg')
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([cat],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```

```python
cat = cv2.imread('cat.jpg',0)
equ = cv2.equalizeHist(cat)
res = np.hstack((cat,equ))
plt.imshow(res,cmap='gray')
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# 2D卷积
img = cv2.imread('cat.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

kernel = np.ones((5,5),np.float32)/25
dst = cv2.filter2D(img,-1,kernel)

plt.subplot(121)
plt.imshow(img)
plt.title('Original')
plt.subplot(122)
plt.imshow(dst)
plt.title('Averaging')
plt.show()
```

```python
# 均值滤波
blur = cv2.blur(img,(5,5))
plt.imshow(blur)
plt.show()
```

```python
# 高斯滤波
blur = cv2.GaussianBlur(img,(5,5),0)
plt.imshow(blur)
plt.show()
```

```python
# 中值滤波
median = cv2.medianBlur(img,5)
plt.imshow(median)
plt.show()
```

```python
# 双边滤波
blur = cv2.bilateralFilter(img,9,75,75)
plt.imshow(blur)
plt.show()
```

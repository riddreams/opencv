```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
cat = cv2.imread('cat.jpg')
cat = cv2.cvtColor(cat,cv2.COLOR_BGR2RGB)
plt.imshow(cat)
plt.show()
```

```python
height,width = cat.shape[:2]
res = cv2.resize(cat,(2*width,2*height),interpolation=cv2.INTER_CUBIC)
#res = cv2.resize(cat,None,fx=2,fy=2,interpolation=cv2.INTER_CUBIC)
plt.imshow(res)
plt.show()
```

```python
M = np.array([[1,0,50],[0,1,30]],np.float32)
res2 = cv2.warpAffine(cat,M,(width,height))
plt.imshow(res2)
plt.show()
```
